```python
import pandas as pd
import genetic_algorithm as ga
```

#  Dataset with 15 cities (p01)

```python
nodes = pd.read_csv("https://people.sc.fsu.edu/~jburkardt/datasets/tsp/p01_xy.txt",sep="\s+",header=None)
nodes = nodes.rename(columns={0:"X",1:"Y"})
nodes = nodes.astype(float)
```

```python
hist, fitness, route, _ = ga.genetic_algorithm(1000,nodes,40,0.1)
ga.plot_results(hist, route, fitness, nodes)
```

# Dataset Berlin52

```python
berlin52 = pd.read_csv("http://elib.zib.de/pub/mp-testdata/tsp/tsplib/tsp/berlin52.tsp",skiprows=7,header=None)
berlin52 = berlin52[0].str.split(" ",expand=True)[[1,2]]
berlin52 = berlin52.rename(columns={1:"X",2:"Y"})
berlin52 = berlin52.iloc[:-1,:]
berlin52 = berlin52.astype(float)
```

```python
hist, fitness, route, _ = ga.genetic_algorithm(200,berlin52,2000,0.25)
ga.plot_results(hist, route, fitness, berlin52)
```
